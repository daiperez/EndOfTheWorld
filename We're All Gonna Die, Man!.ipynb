# Introduction <a id='ex0'></a>

The concept of the zombie has infatuated humans for hundreds of years, with the first mention of zombies appearing in Haiti in the early 17th century[1][4]. The Haitian zombie was the result of voodoo and black magic acting to reanimate a corpse[4]. It is also hypothesized that the ancient Greeks shared a similar fear for the reanimation of corpses as evidenced by attaching weights and bricks to the dead, however largely focused on vampires[7].

The modern idea of the zombie is often the subject of horror and thriller movies and video games[5]. The living dead are mindless beasts that threaten the survival of the human race in its entirety. Zombies are speculated to be  the result of a transmissible disease spread via infectious material contacting human tissue, typically through the advent of a zombie catching a human. The spread of the zombie infection also reaches into graveyards, where the undead are reanimated to feast upon the living.

The SIR model, or Susceptible, Infected, Recovered model is often used to model the spread of infectious diseases in a given population[2]. The SIR model was modified to replace the recovered population with both a zombie and dead population[4]. The susceptible population is the initial population of humans. As the disease begins to infect individuals, they are moved into the infected group. After a specified amount of time as an infected individual, the zombie virus will overtake the human, causing them to turn into a zombie. 

The SIR model can be represented by sets of ordinary differential equations, or ODEs. The ODEs utilize parameters such as the infection rate, zombie death rate, initial population sizes, if the population is quarantining infected individuals, etc, to calculate the spread of the infection.However, since there is no known recovery from the zombie infection, the SIR model was modified to instead calculate the number of zombies and total number of deaths in a population[6].

The fourth order Runge-Kutta method was utilized to approximate the values for these differential equations[3]. The fourth order Runge-Kutta method is widely used due to its accuracy and efficiency. The approximations of these populations as a function of time are used to generate graphs that display how the human race may fare given a specific set of parameters during a simulation.



# Model<a id='ex1'></a>
The modified SIR model for infectious disease[2] is represented by the following ODEs.

\begin{equation*}    
S'(t) = h - V*β*S_{i}*Z_{i} - ΔS*S_{i} \tag {1}
\end{equation*}

\begin{equation*}    
I'(t) = V*β*S_{i}*Z_{i} - ρ*I_{i} - ΔI*I_{i} \tag {2}
\end{equation*}

\begin{equation*}    
Z'(t) = ρ*I_{i} -  α*S_{i}*Z_{i} \tag {3}
\end{equation*}

\begin{equation*}    
D'(t) = ΔS*S_{i} + ΔI*I_{i} + α*S_{i}*Z_{i} \tag {4}
\end{equation*}
          
\begin{equation*}    
Q'(t) = k*I_{i} + σ*Z_{i} - γ*Q_{i}  \tag {5}
\end{equation*}
    
Equation 1 represents the number of humans who are susceptible to infection by the zombie virus. The variable h represents the original number of non-infected humans in the system. The β * S * z factor calculates the number of humans that have been infected with the zombie virus. ΔS*S calculates the change in the susceptible population as a result of human death.

Equation 2 calculates the number of humans who are infected with the zombie virus. This population depends on the changes to the susceptible population.

Finally, equation 4 is made to calculate the total number of dead humans and zombies.

The SIR model also typically includes an additional differential equation, R'(t), that represents the recovered population. Unfortunately due to the nature of a zombie apocalypse there will be no recovery. The recovery equation has been replaced by the death toll, D'(t).

A master list of variables for the above differential equations:

$S_i$ - Initial number of susceptible humans.

$Z_i$ - Initial number of zombies.

$I_i$ - Initial number of infected humans.

$D_i$ - Initial number of dead humans.

$Q_i$ - Initial number of quarantined humans.

$V$ - Zombie to human velocity ratio.

$h$ - Increase in human population per unit time.

$β$ - Probability of human being infected with the zombie virus per unit time.

$ΔS$ - Probability of a human death per unit time.

$ΔI$ - Probability of an infected human death per unit time.

$ρ$ - Probability of infected human succumbing to zombie virus per unit time.

$G$ - Probability of a dead human becoming a zombie per unit time.

$k$ - Rate of infected human quarantine per unit time.

$α$ - Probability of zombie death per unit time.

$σ$ - Rate of zombie quarantine per unit time.

$γ$ - Probability of human attempting to escape quarantine per unit time.

# Numerical Methods<a id='ex2'></a>

This code uses a modified version of the fourth order Runge-Kutta methods outlined in Mark Newman's Computational Physics. This method was selected due to its high accuracy and efficiency over both the Euler and lower order Runge-Kutta methods.  The fourth order Runge-Kutta method uses the following system of equations to solve ODEs by repeatedly estimating the slopes of the functions at a given position and time. 

\begin{equation*}    
k_{1} = hf(x,t) \tag {5}
\end{equation*}

\begin{equation*}    
k_{2} = hf(x+0.5k_{1}, t+0.5h) \tag {6}
\end{equation*}

\begin{equation*}    
k_{3} = hf(x+0.5k_{2},t+0.5h) \tag {7}
\end{equation*}

\begin{equation*}    
k_{4} = hf(x+k_{3}, t+h) \tag {8}
\end{equation*}

\begin{equation*}    
x(t+h) = x(t)+\frac{1}{6}(k_{1}+2k_{2}+2k_{3}+k_{4}) \tag {9}
\end{equation*}

Where the k(1-4) are calculated slopes of the functions at specified intervals, and h is the size of a step inside of a defined interval. These equations are defined in the below cell in the ODE solver function. This function has been edited from the original Mark Newman solve function to be able to calculate the susceptible, infected, zombie, dead, and quarantined units in the system.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.ion()
plt.rcParams['figure.figsize'] = 10, 8


def solve(f,y0,interval,steps,order):
    """ Solve ODE by Euler or Runge-Kutta methods, with fixed number
    of steps.

    In contrast to the examples of Newman Chapter 8, which build up a
    list, point by point, 
    
    f: function giving ODE as y'=f(x,y)
    y0: initial value
    interval: tuple region (a,b) on which to solve ODE
    steps: number of steps
    order: order of solution method (1 for Euler, 2 or 4 for Runge-Kutta)
    
    Returns (x,y) points, as (steps+1)x2 numpy array.
    """
    (a,b)=interval
    temp = 0
    h1=(b - a) / float(steps)
    x=y0
    tpoints= np.arange(a, b, h1)
    spoints = []
    ipoints = []
    zpoints = []
    dpoints = []
    qpoints = []
    
    if (order== 1):
        #Using Euler's method
        for t in (tpoints):
            xpoints.append(x)
            x += h1 * f(x, t)
        func = (tpoints, xpoints)
    if (order == 2):
        for t in (tpoints):
            xpoints.append(x)
            k1 = h1* f(x,t)
            k2 = h1* f(x+0.5*k1, t+0.5*h1)
            x+= k2
        func = (tpoints, xpoints)
    if (order == 4):
        for t in tpoints:
            spoints.append(x[0])
            ipoints.append(x[1])
            zpoints.append(x[2])
            dpoints.append(x[3])
            qpoints.append(x[4])
            k1 = h1*f(x,t)
            k2 = h1*f(x+0.5*k1,t+0.5*h1)
            k3 = h1*f(x+0.5*k2,t+0.5*h1)
            k4 = h1*f(x+k3,t+h1)
            x += (k1+2*k2+2*k3+k4)/6

    return (tpoints, spoints, ipoints, zpoints, dpoints, qpoints)
```

```python
def f(y, t):
    '''Declares all of my functions that are being used in the code
    
    f = a function giving an ODE as y'=f(x,y).  '''
    #initialize
    Si=y[0]
    Ii=y[1]
    Zi=y[2]
    Di=y[3]
    Qi=y[4]
    

    f0= h - V*beta*Si*Zi - delta_S*Si 
    f1= V*beta*Si*Zi - ro*Ii - delta_I*Ii-k*Ii
    f2= ro*Ii - alpha*Si*Zi + G*Di-sigma*Zi
    f3= delta_S*Si + delta_I*Ii + alpha*Si*Zi - G*Di+ gamma*Qi
    f4= k*Ii + sigma*Zi - gamma*Qi
    
    return np.array([f0, f1, f2, f3, f4], float)
```

## Zombie Apocalypse<a id='ex4'></a>

```python
#declare range
interval= [0,8]
steps= 1000

h=20.            #rate at which people coming into the area/birth rate
beta=0.03        #transmission rate
ro =1            #percent of those who are infected become zombies
alpha= 0.2*beta  #probability that human kills zombie
delta_I= 0.0     #probability that an infected human is killed/dies
delta_S= 0.0     #Probability that a human is killed/dies
G=0.06           #probability that a dead individual turns to zombie

#Since we are not taking quarantine into account just yet, they should be 0
k=0.             #rate at which infected people get quarantined
sigma=0.         #rate at which zombie population get quarantined
gamma=0.         #The chance that some members will try to escape and get killed
V=1              #velocity of zombies.


#initial conditions for the population

S0 = 500.        #initial Population
I0=0.            #initial infected population
Z0=1.            #initial zombie population
D0=0.            #initial deaths
Q0=0

y0=[S0, I0, Z0, D0, Q0]

(tp, sp, ip, zp, dp, qp)= solve(f, y0, interval, steps, order=4)
```
